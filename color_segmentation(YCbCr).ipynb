```python
def RGB2YCbCr (RGB):
    transform = np.array([[.299, .587, .114], [.5, -.4187, -.0813], [-.1687, -.3313, .5]])
    ycbcr = np.dot(RGB, np.transpose(transform))
    ycbcr[:,[1,2]] += 128
    return np.uint8(ycbcr)
```

```python
import numpy as np
import os

# find parameters
path = "npz/"
path_list = os.listdir(path)
path_list.sort()
data = [[]] * 200
i = 0
for filename in path_list:
    #print(os.path.join(path,filename))
    data[i] = np.load(os.path.join(path,filename))
    i = i + 1

n = i

blue = [[0, 0, 0]]
brwn = [[0, 0, 0]]
orng = [[0, 0, 0]]
othr = [[0, 0, 0]]
ored = [[0, 0, 0]]
sred = [[0, 0, 0]]
    
for i in range(0, n):
    if (data[i]['COLOR_BLUE.npy'].size != 0):
        blue = np.concatenate((blue, data[i]['COLOR_BLUE.npy']), axis = 0)
    if (data[i]['COLOR_BROWN.npy'].size != 0):
        brwn = np.concatenate((brwn, data[i]['COLOR_BROWN.npy']), axis = 0)
    if (data[i]['COLOR_ORANGE.npy'].size != 0):
        orng = np.concatenate((orng, data[i]['COLOR_ORANGE.npy']), axis = 0)
    if (data[i]['COLOR_OTHER.npy'].size != 0):
        othr = np.concatenate((othr, data[i]['COLOR_OTHER.npy']), axis = 0)
    if (data[i]['COLOR_OTHER_RED.npy'].size != 0):
        ored = np.concatenate((ored, data[i]['COLOR_OTHER_RED.npy']), axis = 0)
    if (data[i]['COLOR_STOP_SIGN_RED.npy'].size != 0):
        sred = np.concatenate((sred, data[i]['COLOR_STOP_SIGN_RED.npy']), axis = 0)

blue = blue[1:len(blue),:]
brwn = brwn[1:len(brwn),:]
orng = orng[1:len(orng),:]
othr = othr[1:len(othr),:]
ored = ored[1:len(ored),:]
sred = sred[1:len(sred),:]

# prior
total = len(blue) + len(brwn) + len(orng) + len(othr) + len(ored) + len(sred)
prior_blue = len(blue) / total
prior_brwn = len(brwn) / total
prior_orng = len(orng) / total
prior_othr = len(othr) / total
prior_ored = len(ored) / total
prior_sred = len(sred) / total

# RGB
blue_mu = np.mean(blue, axis = 0)
brwn_mu = np.mean(brwn, axis = 0)
orng_mu = np.mean(orng, axis = 0)
othr_mu = np.mean(othr, axis = 0)
ored_mu = np.mean(ored, axis = 0)
sred_mu = np.mean(sred, axis = 0)

blue_cov = np.cov(np.transpose(blue))
brwn_cov = np.cov(np.transpose(brwn))
orng_cov = np.cov(np.transpose(orng))
othr_cov = np.cov(np.transpose(othr))
ored_cov = np.cov(np.transpose(ored))
sred_cov = np.cov(np.transpose(sred))
```

```python
# single gaussian

# slower
def gaussian_(x, mu, sig):
    cov_inv = np.linalg.inv(sig)
    denominator = np.sqrt((2*np.pi)**3 * np.linalg.det(sig))
    exp = [0]*len(x)
    
    for i in range(len(x)):
        exp[i] = np.exp(-(np.dot(np.dot(x[i]-mu,cov_inv), np.transpose(x[i]-mu))) / 2)
        
    return exp / denominator

# faster
def gaussian(x, mu, sig):
    L = np.linalg.cholesky(np.linalg.inv(sig))
    expTerm = np.exp(-np.sum(np.square(np.dot(x-mu, L)), axis = 1) / 2)
    denominator = np.sqrt((2*np.pi)**3 * np.linalg.det(sig))
    return expTerm / denominator
```

```python
# load image: done
def load_image(imgBGR, color):
    if color == 'RGB':
        img = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
    elif color == 'HSV':
        img = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2HSV)
    else:
        img = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2YCR_CB)

    pixels_array = np.reshape(img, (-1,3))
    return pixels_array
```

```python
from skimage.measure import label, regionprops

class StopSignDetector():
    def __init__(self):
        self.prior_blue = prior_blue
        self.prior_brwn = prior_brwn
        self.prior_orng = prior_orng
        self.prior_othr = prior_othr
        self.prior_ored = prior_ored
        self.prior_sred = prior_sred
            
        if color == 'RGB':
            self.mu_blue = np.array(blue_mu)
            self.mu_brwn = np.array(brwn_mu)
            self.mu_orng = np.array(orng_mu)
            self.mu_othr = np.array(othr_mu)
            self.mu_ored = np.array(ored_mu)
            self.mu_sred = np.array(sred_mu)
            
            self.cov_blue = np.array(blue_cov)
            self.cov_brwn = np.array(brwn_cov)
            self.cov_orng = np.array(orng_cov)
            self.cov_othr = np.array(othr_cov)
            self.cov_ored = np.array(ored_cov)
            self.cov_sred = np.array(sred_cov)
        
        elif color == 'YCbCr':
            self.mu_blue = np.array(blue_mu_YCbCr)
            self.mu_brwn = np.array(brwn_mu_YCbCr)
            self.mu_orng = np.array(orng_mu_YCbCr)
            self.mu_othr = np.array(othr_mu_YCbCr)
            self.mu_ored = np.array(ored_mu_YCbCr)
            self.mu_sred = np.array(sred_mu_YCbCr)
            
            self.cov_blue = np.array(blue_cov_YCbCr)
            self.cov_brwn = np.array(brwn_cov_YCbCr)
            self.cov_orng = np.array(orng_cov_YCbCr)
            self.cov_othr = np.array(othr_cov_YCbCr)
            self.cov_ored = np.array(ored_cov_YCbCr)
            self.cov_sred = np.array(sred_cov_YCbCr)

    def segment_image(self, img):
        
        image_size = np.shape(img)
        test = load_image(img, color)
        
        # claculate gaussian probability
        p_blue = gaussian(test, self.mu_blue, self.cov_blue) * prior_blue
        p_brwn = gaussian(test, self.mu_brwn, self.cov_brwn) * prior_brwn
        p_orng = gaussian(test, self.mu_orng, self.cov_orng) * prior_orng
        p_othr = gaussian(test, self.mu_othr, self.cov_othr) * prior_othr
        p_ored = gaussian(test, self.mu_ored, self.cov_ored) * prior_ored
        p_sred = gaussian(test, self.mu_sred, self.cov_sred) * prior_sred
        
        # choose the max and output the mask
        max_p = list(map(max, zip(p_blue, p_brwn, p_orng, p_othr, p_ored, p_sred)))
        mask_img = 255*np.array(max_p == p_sred).astype('uint8')
        mask_img = mask_img.reshape(image_size[:2])
        
        # Closing (morphology), erode first and dilate
        kernel = np.ones((10,10), np.uint8)
        mask_img = cv2.morphologyEx(mask_img, cv2.MORPH_CLOSE, kernel)
        
        # Opening (morphology), erode first and dilate
        kernel = np.ones((4,4), np.uint8)
        mask_img = cv2.morphologyEx(mask_img, cv2.MORPH_OPEN, kernel)
        
        #raise NotImplementedError
        return mask_img

    def get_bounding_box(self, img):
        image_size = np.shape(img)
        mask_img = self.segment_image(img)
        
        boxes = []
        labels = label(mask_img)
        regions = regionprops(labels)
        for r in regions:
            score = 0
            minr, minc, maxr, maxc = r.bbox
            h = maxr - minr
            w = maxc - minc
            a = r.area
            e = r.extent
            er = r.euler_number
            
            if h/w > 0.5 and h/w < 2:
                score = score + 2
            if h/w > 0.9 and h/w < 1.2:
                score = score + 3
             
            if e > 0.45 and e < 0.85:
                score = score + 3
            if a > 9000 and e > 0.55 and e < 0.75:
                score = score + 3   
            if a > 500 and a < 9000 and e > 0.72 and e < 0.85:
                score = score + 3
            
            if er < 0:
                score = score + 5;
            if score > 10:
                boxes.append([minc, image_size[0]-maxr, maxc, image_size[0]-minr])
            
#             print("area :", a)
#             print("e    :", e)
#             print("er   :", er)
#             print("score:", score)
        
        sorted(boxes, key = lambda x:(x[0]))
        
        return boxes
     
```

```python
# main function

import numpy as np
import os, cv2

color = 'YCbCr'
folder = "validation"
my_detector = StopSignDetector()
for filename in os.listdir(folder):
    image = cv2.imread(os.path.join(folder,filename))
    
    # resize image
    if image.shape[1] > image.shape[0]:
        w = 1000;
        h = int(image.shape[0] * 1000 / image.shape[1])
    else:
        h = 1000;
        w = int(image.shape[1] * 1000 / image.shape[0])
        
    dim = (w, h)
    img = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
    
    # cv2.imshow('image', img)
    # cv2.waitKey(1)
    # cv2.destroyAllWindows()
    mask_img = my_detector.segment_image(img)
    boxes = my_detector.get_bounding_box(img)

    # print(boxes)
    # print(filename)
    # print("-----------------")

    image_size = np.shape(img)
    box_img = []
    for i in range(len(boxes)):
        box_img = cv2.rectangle(img, (boxes[i][0],image_size[0]-boxes[i][1]), (boxes[i][2],image_size[0]-boxes[i][3]), (0, 255, 0) , 2)

    cv2.imshow('mask', mask_img)
    cv2.waitKey(3)
    cv2.imwrite(os.path.join("result_mask", filename), mask_img)

    if len(box_img) > 0:
        cv2.imshow('box', box_img)
        cv2.waitKey(1)
        cv2.imwrite(os.path.join("result_box", filename), box_img)
    else:
        cv2.imwrite(os.path.join("result_box", filename), image)
    
cv2.waitKey(1)
cv2.destroyAllWindows()
```
